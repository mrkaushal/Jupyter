```python
import spacy
```

```python
nlp = spacy.load('en_core_web_lg')
```

```python
from spacy.matcher import Matcher
```

```python
matcher = Matcher(nlp.vocab)
```

```python
# SolarPower
pattern1 = [{'LOWER': 'solarpower'}]
# Solar-power
pattern2 = [{'LOWER': 'solar'},{'IS_PUNCT': True}, {'LOWER': 'power'}]
# Solar power
pattern3 = [{'LOWER': 'solar'}, {'LOWER': 'power'}]
```

```python
matcher.add('SolarPower',[pattern1,pattern2,pattern3])
```

```python
doc = nlp(u"The Solar Power industry continues to grow a solarpower increases. Solar-power is amazing.")
```

```python
found_matcher = matcher(doc)
```

```python
print(found_matcher)
```

```python
for match_id, start, end in found_matcher:
    string_id = nlp.vocab.strings[match_id]
    span = doc[start: end]
    print(match_id, string_id, start, end, span)
```

```python
matcher.remove('SolarPower')
```

```python
# SolarPower
pattern1 = [{'LOWER': 'solarpower'}]
# Solar-power
pattern2 = [{'LOWER': 'solar'},{'IS_PUNCT': True, 'OP':'*'}, {'LOWER': 'power'}]
# Solar power
pattern3 = [{'LOWER': 'solar'}, {'LOWER': 'power'}]
```

```python
matcher.add('SolarPower', [pattern1,pattern2,pattern3])
```

```python
doc2 = nlp(u"Solar--Power is solarpower yay!")
```

```python
found_matcher = matcher(doc2)
```
