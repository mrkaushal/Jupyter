##### Lemmatization

```python
import spacy
```

```python
nlp = spacy.load('en_core_web_lg')
```

```python
doc1 = nlp(u"I am a runner running in a race because I love to run.")
```

```python
for token in doc1:
    print(token.text, '\t', token.pos_, '\t', token.lemma, '\t', token.lemma_)
```

```python
def show_lemmas(text):
    for token in text:
        print(f'{token.text:{12}}\t{token.pos_:{6}}\t{token.lemma:{22}}\t{token.lemma_}')
```

```python
doc2 = nlp(u"I saw ten mice in the forest.")
```

```python
show_lemmas(doc2)
```
