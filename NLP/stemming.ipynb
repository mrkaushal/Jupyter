##### Stemming

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
p_stemmer = PorterStemmer()
```

```python
words = ['running', 'runs', 'run', 'rans', 'ran', 'easily', 'fairly', 'fairness', 'fair', 'fairly', 'fairness']
```

```python
for word in words:
    print(word + ' => ' + p_stemmer.stem(word))
```

```python
from nltk.stem.snowball import SnowballStemmer
```

```python
s_stemmer = SnowballStemmer(language='english')
```

```python
for word in words:
    print(word + ' => ' + s_stemmer.stem(word))
```

```python
words = ['generous', 'generations', 'generously', 'generate']
```

```python
for word in words:
    print(word + ' => ' + s_stemmer.stem(word))
```
