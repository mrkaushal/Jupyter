```python
library = [('Author', 'Topic', 'Pages'), ('John', 'Python', 100), ('Jane', 'Python', 200)]
```

```python
library
```

```python
for book in library:
    print(f"Author is {book[0]} and topic is {book[1]} and pages are {book[2]}")
```

```python
for author, topic, pages in library:
    print(f"Author is {author} and topic is {topic} and pages are {pages}")
```

```python
for author, topic, pages in library:
    print(f"Author is {author:{10}} {topic:{30}} {pages:->{10}}")
```

```python
from datetime import datetime
```

```python
today = datetime(year=2002, month=6, day=13, hour=12, minute=30, second=45)
```

```python
print(f"Today is {today:%B %d, %Y}")
print(f"Time is {today:%H:%M:%S}")
```
