##### Tokenization

```python
import spacy
nlp = spacy.load('en_core_web_lg')
```

```python
mystring = '"We\'re moving to L.A.!"'
```

```python
print(mystring)
```

```python
doc = nlp(mystring)
```

```python
for token in doc:
    print(token.text)
```

```python
doc8 = nlp(u"Apple to build a Hong Kong factory for $1.6bn")
```

```python
for token in doc8:
    print(token.text)
```

```python
for token in doc8:
    print(token.text, end='||')
```

```python
for entity in doc8.ents:
    print(entity)
    print(entity.label_)
    print(str(spacy.explain(entity.label_)))
    print('\n')
```

```python
doc9 = nlp(u"Autonomous cars shift insurance liability toward manufacturers.")
```

```python
for chunk in doc9.noun_chunks:
    print(chunk.text)
```
